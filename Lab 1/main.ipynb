```python
import numpy as np

def f1(x):
    return -x * np.exp(-x / 2)

def f2(x):
    return np.sin(np.pi * x) + 2 * np.cos(2 * np.pi * x) + 3 * np.sin(2 * np.pi * x) * np.exp(-x / 2)

def f3(x):
    return 2 * x * np.exp(-x)

x = np.linspace(0, 10, 1000)

def f_length(f, x):
    dx = np.diff(x)
    
    y = f(x)

    dy = np.diff(y)
    
    length = np.sum(np.sqrt(dx**2 + dy**2))
    
    return length


f1_length = f_length(f1, x)
f2_length = f_length(f2, x)
f3_length = f_length(f3, x)


print(f"Długość krzywej dla f1(x) w przedziale [0, 10]: {f1_length}")
print(f"Długość krzywej dla f2(x) w przedziale [0, 10]: {f2_length}")
print(f"Długość krzywej dla f3(x) w przedziale [0, 10]: {f3_length}")
```

```python
from matplotlib import cm
from matplotlib.ticker import LinearLocator

x=np.linspace(-5,5,100)
y=np.linspace(-5,5,100)
x,y =np.meshgrid(x,y)
function= np.sin(np.sqrt(2 * x**2 + y**2))
z=function

fig= plt.figure(figsize=(14,7))

ax1=fig.add_subplot(121, projection='3d')
surface = ax1.plot_surface(x,y,z, cmap=cm.coolwarm)
ax1.set_zlim(-1.5,1.5)
ax1.zaxis.set_major_locator(LinearLocator(10))
ax1.zaxis.set_major_formatter('{x:.02f}')
ax1.set_title("Wykres powierzchniowy")
ax1.set_xlabel("X")
ax1.set_ylabel("Y")


ax2=fig.add_subplot(122)
contour= ax2.contour(x,y,z, cmap=cm.coolwarm)
ax2.set_title("Wykres konturowy")
ax2.set_xlabel("X")
ax2.set_ylabel("Y")

fig.colorbar(surface, shrink=0.5, aspect=5)
fig.colorbar(contour, shrink=0.5, aspect=5)


```
